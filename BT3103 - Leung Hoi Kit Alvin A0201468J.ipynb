```python
import pandas as pd
```

```python
df = pd.read_csv('FIFA.csv')
df
```

```python
import matplotlib.pyplot as plt
```

```python
plt.hist(df['Rating'])
plt.xlabel('Rating')
plt.show()
```

```python
plt.hist(df['Age'])
plt.xlabel('Age')
plt.show()
```

```python
plt.boxplot(df['Age'])
plt.show()
```

```python
plt.figure(figsize=(20,10))
plt.scatter(x=df["Speed"],y=df["Acceleration"],s=50,c=df['Age'])
plt.xlabel('Speed');plt.ylabel('Acceleration')
plt.show()
```

```python
nationality = df['Nationality'].value_counts().head(8)
x=nationality.index
y=nationality.values
plt.bar(x,y)
plt.show()
```

```python
plt.plot(x,y)
plt.show()
```

```python
import seaborn as sns
```

```python
df_mpg = sns.load_dataset("mpg")
df_mpg
```

```python
sns.relplot(x='horsepower',y='mpg',data=df_mpg,kind='line')
```

```python
sns.relplot(x='horsepower',y='mpg',data=df_mpg,hue='origin')
```

```python
sns.catplot(x='horsepower',y='mpg',data=df_mpg,kind='swarm')
```

```python
sns.catplot(x='horsepower',y='mpg',data=df_mpg,hue='origin')
```

```python
sns.catplot(x='cylinders',y='mpg',data=df_mpg,hue='origin',kind='box')
```

```python
sns.catplot(x='origin',y='mpg',data=df_mpg,hue='cylinders',kind='bar')
```
